### Importation of neccessary libraries

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()

from sklearn.datasets import fetch_20newsgroups

data = fetch_20newsgroups()
data.target_names
```

```python
# Defining all categories
categories = ['alt.atheism', 'comp.graphics', 'comp.os.ms-windows.misc', 'comp.sys.ibm.pc.hardware',
 'comp.sys.mac.hardware', 'comp.windows.x', 'misc.forsale', 'rec.autos', 'rec.motorcycles',
 'rec.sport.baseball', 'rec.sport.hockey', 'sci.crypt', 'sci.electronics', 'sci.med', 'sci.space',
 'soc.religion.christian', 'talk.politics.guns', 'talk.politics.mideast', 'talk.politics.misc',
 'talk.religion.misc']


# Train the data on theese categories
train = fetch_20newsgroups(subset = "train", categories = categories)

# Testing the data for these categories
test = fetch_20newsgroups(subset = "test", categories = categories)

# Printing training data
print(train.data[5])
```

### Actual Prediction

```python
# Importing neccessary packages

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline
```

```python
# Creating model based on Multinomial Naive Bayes
model = make_pipeline(TfidfVectorizer(), MultinomialNB())
```

```python
# Training the model
model.fit(train.data, train.target)
```

```python
# Creating labels for the dataset
label = model.predict(test.data)
```

### Creating confusion matrix and Heatmap

```python
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(test.target, label)
sns.heatmap(mat.T, square = True, annot = True, fmt = "d", cbar = False, xticklabels = train.target_names, yticklabels = train.target_names)

#Plotting Heatmap of confusion matrix
plt.xlabel("True Label")
plt.ylabel("Predicted Label")
```

### Testing the Prediction

```python
# Predicting categories on new data based on the trained model

def predict_category(s, train = train, model = model):
    pred = model.predict([s])
    return train.target_names[pred[0]]
```

```python
predict_category("Jesus Christ")
```

```python
predict_category("Sending load to internation space station")
```

```python
predict_category("BMW is better that Toyota")
```
